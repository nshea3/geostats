```
import pandas as pd
import numpy as np
```

```
ip_txtfile = 'https://raw.githubusercontent.com/nshea3/geostats/master/data/barbour_county_wv_initial_potential.txt'
#East (Km)           
#North (Km)               
#Initial Potential (Mcfpd)
ip_df = pd.read_fwf(ip_txtfile, sep=' ', skiprows=5, names=['East', 'North', 'Initial_Potential'])
ip_df.East = pd.to_numeric(ip_df.East)
ip_df.North = pd.to_numeric(ip_df.North)
ip_df.Initial_Potential = pd.to_numeric(ip_df.Initial_Potential)
ip_df.head()
```

```
ip_df.plot.scatter(x='East', y='North', c='Initial_Potential')
```
